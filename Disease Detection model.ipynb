# Using Naiv Bayes Algorithm of supervised learning

```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
data  = pd.read_csv("training.csv")
data.head()
```

```python
x = data.iloc[:,:-2].values
y = data['prognosis'].values
```

```python
#spliting the test train data
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=35)
```

```python
from sklearn.preprocessing import StandardScaler
dtc = StandardScaler()
xtrain= dtc.fit_transform(xtrain)
xtest = dtc.transform(xtest)
```

```python
from sklearn.naive_bayes import GaussianNB
gs = GaussianNB()
model= gs.fit(xtrain,ytrain)
model
```

```python
y_pred = model.predict(xtest)
y_pred
```

```python
from sklearn import metrics
acc = metrics.accuracy_score(y_pred,ytest)
acc
```

```python
#printing the confusion matrics
from sklearn.metrics import confusion_matrix
con = metrics.confusion_matrix(y_pred,ytest)
print(con)
```

```python
#calculating the accuracy from confusion matrix
a = con.shape
correct_count=0
false_count= 0

for row in range(a[0]):
    for col in range(a[1]):
        if row==col:
            correct_count+=con[row,col]
        else:
            false_count+=con[row,col]
total_count = correct_count+false_count
print("Correct Prediction",correct_count)
print("Wrong Prediction",false_count)
print("Accuracy Score ",correct_count/total_count)
```

```python
testData= pd.read_csv("testing.csv")
xt = testData.iloc[:,:-1].values
yt = testData['prognosis'].values

testresult= model.predict(xt)
testresult

```

```python
#finding accuracy of the test data
ac = metrics.accuracy_score(yt,testresult)
ac
```
